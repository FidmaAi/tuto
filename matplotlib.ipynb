```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Some of the most commonly used modules in Matplotlib are:

1. **pyplot**:
   - **Description**: This module is the core of Matplotlib's plotting capabilities and is often the only one many users directly interact with. It provides a MATLAB-like interface for creating figures and plots.
   - **Common Functions**:
     - `plt.plot()`: Plot lines and/or markers.
     - `plt.scatter()`: Create scatter plots.
     - `plt.bar()`: Create bar charts.
     - `plt.hist()`: Create histograms.
     - `plt.xlabel()`, `plt.ylabel()`: Set the labels of the x and y axes.
     - `plt.title()`: Set the title of the plot.
     - `plt.legend()`: Add a legend to the plot.
     - `plt.show()`: Display the plot.

2. **Figure**:
   - **Description**: This module deals with the figure object, which is the overall container for all plot elements.
   - **Common Methods**:
     - `Figure()`: Create a new figure.
     - `fig.add_subplot()`: Add a subplot to the figure.
     - `fig.savefig()`: Save the figure to a file.

3. **Axes**:
   - **Description**: This module contains the axes object, which is the area where data is plotted. Each figure can contain multiple axes.
   - **Common Methods**:
     - `ax.plot()`: Plot lines and/or markers on the axes.
     - `ax.scatter()`: Create scatter plots on the axes.
     - `ax.bar()`: Create bar charts on the axes.
     - `ax.hist()`: Create histograms on the axes.
     - `ax.set_xlabel()`, `ax.set_ylabel()`: Set the labels of the x and y axes.
     - `ax.set_title()`: Set the title of the axes.
     - `ax.legend()`: Add a legend to the axes.

4. **Artist**:
   - **Description**: This module is the base class for all elements that can be drawn on a figure.
   - **Common Subclasses**:
     - `Line2D`: Represents a line.
     - `Patch`: Represents shapes like rectangles and circles.
     - `Text`: Represents text.
     - `Collection`: Represents a collection of objects (e.g., a collection of lines or patches).

5. **Animation**:
   - **Description**: This module provides classes for creating animations.
   - **Common Classes**:
     - `FuncAnimation`: Creates an animation by repeatedly calling a function.
     - `ArtistAnimation`: Creates an animation by drawing a series of Artist objects.

6. **Colors**:
   - **Description**: This module provides functions for handling colors.
   - **Common Functions**:
     - `to_rgb()`: Convert a color to an RGB tuple.
     - `to_hex()`: Convert a color to a hexadecimal string.
     - `cnames`: A dictionary of color names to RGB values.

7. **cm (colormaps)**:
   - **Description**: This module provides functions and classes for handling colormaps.
   - **Common Colormaps**:
     - `viridis`, `plasma`, `inferno`, `magma`: Popular perceptually uniform colormaps.
     - `jet`: A widely used but non-uniform colormap.

8. **gridspec**:
   - **Description**: This module provides a way to layout subplots in a grid.
   - **Common Classes**:
     - `GridSpec`: Specifies the geometry of the grid that a subplot will be placed into.
     - `SubplotSpec`: Specifies the location of a subplot in a GridSpec layout.

These modules and their associated functions and classes form the backbone of Matplotlib's functionality, allowing users to create a wide variety of static, animated, and interactive plots.

```python
plt.figure(figsize=(5,3), dpi=100)
plt.plot(np.arange(20),np.arange(100,120), label="2x",color="green",linewidth=3 ,linestyle="dashed", marker=".",markersize=5, markeredgecolor="red")
#line number2 

x2=np.arange(0,20,0.5)
plt.plot(x2,x2**2)
plt.title("random ",fontdict={"fontsize":20})
plt.xlabel('1st array ',fontdict={"fontname":"Comic Sans MS"})
plt.ylabel('1st array ')
plt.xticks()
plt.yticks()
plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```
